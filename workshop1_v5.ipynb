## Описание и загрузка данных


* Date: Год выпуска автомобиля.
* Make: Марка автомобиля.
* Model: издание автомобиля определенной марки.
* Trim: Уровни отделки салона автомобиля — это просто разные версии модели.
* Body: Тип кузова транспортного средства относится к форме и модели конкретной марки автомобиля.
* Transmission: механизм, который передает мощность от двигателя к колесам.
* VIN: идентификационный номер транспортного средства.
* State: состояние, в котором автомобиль выставлен на аукцион.
* Condition: Состояние автомобилей на момент аукциона.
* Odometer: расстояние, пройденное автомобилем с момента выпуска.
* Color: Цвет кузова автомобиля.
* Interior: Цвет салона автомобиля.
* Seller: Продавец автомобиля, автосалоны.
* mmr: Рекорд рынка Manhiem, рыночная оценочная цена автомобилей.
* sellingprice: цена, по которой автомобиль был продан на аукционе.
* saledate: Дата продажи автомобиля.

#### 4x4

```python
def is_4x4(row):
    trim = str(row['trim'])
    make = row['make']
    pattern = re.compile(r'awd|4x4|quattro|4-matic|xdrive|xdr|xi|4motion|wrx|4s|4matic')
    if re.search(pattern, trim):
        return 'awd'
    elif make in ['subaru', 'jeep','land rover']:
        return 'awd'
    else:
        return 'unknown'
        
    
```

```python
combined['awd'] = combined.apply(is_4x4, axis = 1)
```

```python
combined.groupby('awd')['vin'].count()
```

#### special

```python
def is_special(row):
    trim = str(row['trim'])
    pattern = re.compile(r'luxury|premium|leather|special|premier|executive|edition|design|amg|elite|limited|sti|brabus|package|packages|sport')
    base_pattern = re.compile(r'base|standard')
    if re.search(pattern, trim):
        return 'special'
    elif re.search(base_pattern, trim):
        return 'base'
    else:
        return 'unknown'
```

```python
combined['special'] = combined.apply(is_special, axis = 1)
```

```python
combined.groupby('special')['vin'].count()
```

#### engine

```python
def what_engine(row):
    trim = str(row['trim'])
    v6 = re.compile(r'v6|v-6')
    v8 = re.compile(r'v8|v-8')
    
    if re.search(v6, trim):
        return 'v6'
    elif re.search(v8, trim):
        return 'v8'
    else:
        return 'unknown'
```

```python
combined['engine'] = combined.apply(what_engine, axis = 1)
```

```python
combined.groupby('engine')['vin'].count()
```

#### volume

```python
def volume(trim):
    pattern = r'\d+\.\d+?'
    matches = re.findall(pattern, trim)
    
    if matches:
        try:
            return float(matches[0])
        except ValueError:
            return None
    
    return None
```

```python
volume('3.2 premium ')
```

```python
def get_volume(row):
    trim = str(row['trim'])
    result = volume(trim)
    try:
        if result <10:
            return result
        else:
            return None
    except:
        return None
```

```python
combined['volume'] = combined.apply(get_volume, axis = 1)
```

```python
def add_new_features(data_source):
    data = data_source
    
    print('Starting data preparation...')
    print('Preparing make, make segments and brand price groups')
    
    data['make'] = data['make'].str.lower()
    data['make'] = data['make'].map(car_brands_dict).fillna(combined['make'])
    data['make'] = data['make'].fillna('unknown')
    
    data['make_segment'] = data.apply(make_segment, axis = 1)
    data['brand_price_groups'] = data.apply(brand_price_groups, axis = 1)
    
    print('Preparing year_date, car_age and car_age_groups')
    data['year_date'] = pd.to_datetime(data['year'], format='%Y')
    data['car_age'] = data.apply(car_age, axis = 1)
    data['car_age_group'] = data.apply(car_age_group, axis = 1)
    
    print('Preparing model, trim, awd and special')
    data['model'] = data['model'].str.lower()
    data['model'] = data['model'].fillna('unknown')
    
    data['trim'] = data['trim'].str.lower()
    data['trim'] = data['trim'].fillna('unknown')
    
    data['awd'] = data.apply(is_4x4, axis = 1)
    data['special'] = data.apply(is_special, axis = 1)
    data['engine'] = data.apply(what_engine, axis = 1)
    
    print('Working with bodies')
    data['body'] = data['body'].str.lower()
    data['body'] = data.apply(is_pickup, axis = 1)
    data['body'] = data.apply(is_coupe, axis = 1)
    data['body'] = data.apply(is_van, axis = 1)
    data['body'] = data.apply(split_return, axis = 1)
    data['body'] = data['body'].fillna('unknown')
    
    print('Transmission and state')
    data['transmission'] = data['transmission'].fillna('automatic')
    data['state'] = data['state'].fillna('unknown')
    
    print('Working on condition and condition_groups')
    data['condition'] = data['condition'].fillna(0)
    data['condition_groups'] = data.apply(condition_groups, axis = 1)
    
    print('Now odometer, color, interior types')
    data['odometer'] = data['odometer'].fillna(data['odometer'].median())
    data['color'] = data['color'].replace('—','unknown')
    data['color'] = data['color'].fillna('unknown')
    data['interior'] = data['interior'].replace('—','unknown')
    data['interior'] = data['interior'].fillna('unknown')
    data['interior'] = data.apply(simplify_interior, axis = 1)
    
    print('Seller types')
    data['sold_by_dealer'] = data.apply(sold_by_dealer, axis = 1)
    data['sold_by_bank'] = data.apply(sold_by_bank, axis = 1)
    data['saledate_converted'] = data['saledate'].apply(convert_to_datetime)
    
    print('Sale date and sale month')
    data['sale_day'] = data['saledate_converted'].dt.dayofweek
    data['sale_day'] = data['sale_day'].map(days_map)
    
    data['sale_month'] = data['saledate_converted'].dt.month
    data['sale_month'] = data['sale_month'].map(month_map)
    
    
    data['state'] = data.apply(state_cleaner, axis = 1)
    
    print('Usage intensity and usage groups')
    data['usage'] = data.apply(usage_calculator, axis = 1)
    data['usage_groups'] = pd.qcut(data['usage'], 8)
    data['usage_groups'] = data['usage_groups'].astype(str)
    
    print('Brand popularity and model popularity')
    data['brand_popularity_group'] = data.apply(brand_popularity_groups, axis = 1)
    data['model_popularity_group'] = data.apply(model_popularity_groups, axis = 1)
    
    print('Other variables')
    data['volume'] = data.apply(get_volume, axis = 1)
    data['renter'] = data.apply(sold_by_renter, axis = 1)
    data['leased'] = data.apply(is_leased, axis = 1)
    data['sold_by_any_dealer'] = data.apply(sold_by_any_dealer, axis = 1)
    data['sold_by_any_dealer'] = data['sold_by_any_dealer'].fillna('unknown')
    
    print('All done')
    
    return data
```
